```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('youtube_india_usa.csv')
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='category_name',y='views', hue='country',estimator='mean')
plt.xticks(rotation=45)
plt.title("Average Views by Category (India vs USA)")
plt.xlabel("Category")
plt.ylabel("Average Views")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(data=df, x='country',y='like_ratio')
plt.title("LIke Ratio Distribution by Country")
plt.xlabel("Country")
plt.ylabel("Like Ratio")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df,x='publish_hour',y='views',hue='country',estimator='mean')
plt.title("Best Publishing Hour by Country")
plt.xlabel("Hour of Day")
plt.ylabel("Average Views")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    data=df,
    x='publish_day',
    y='views',
    estimator='mean',
    order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
)
plt.title("Average Views by Publish Day")
plt.ylabel("Average Views")
plt.xlabel("Day")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x='title_length',
    y='views',
    hue='country',
    alpha=0.5
)
plt.title("Title Length vs Views")
plt.xlabel("Title Length (characters)")
plt.ylabel("Views")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    data=df,
    x='category_name',
    y='like_ratio',
    estimator='mean'
)
plt.xticks(rotation=60)
plt.title("Average Like Ratio by Category")
plt.ylabel("Average Like Ratio")
plt.xlabel("Category")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    data=df,
    x='category_name',
    y='days_to_trend'
)
plt.xticks(rotation=60)
plt.title("Days Taken to Trend by Category")
plt.ylabel("Days to Trend")
plt.xlabel("Category")
plt.show()

```
