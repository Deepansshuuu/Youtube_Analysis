```python
import pandas as pd
import numpy as np
```

```python
df_india = pd.read_csv('final_india.csv')
df_usa = pd.read_csv('final_usa.csv')
```

```python
df_india['country'] = 'India'
df_usa['country'] = 'United States'
```

```python
print(df_india.columns)
print(df_usa.columns)
```

```python
df_usa = df_usa[df_india.columns]
```

```python
print(df_india.columns)
print(df_usa.columns)
```

```python
df_all = pd.concat([df_india,df_usa],ignore_index=True)
```

```python
df_all['country'].value_counts()
```

```python
df_all.to_csv('youtube_india_usa.csv',index=False)
```
